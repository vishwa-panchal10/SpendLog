```python
import matplotlib.pyplot as plt 
import pandas as pd 
from datetime import datetime, timedelta 

class ExpenseList: 

    def __init__(self): 
        self.expenses = [] 

    def add_expense(self, date, category, amount): 
        self.expenses.append({'date': date, 'category': category, 'amount': amount}) 

    def visualize_expenses(self): 
        print("\nExpense Visualization\n") 
        print("1. Daily Expenses") 
        print("2. Weekly Expenses") 
        print("3. Monthly Expenses") 
        print("4. Yearly Expenses")
        print("5. Pie Chart") 
        print("6. Tabular View") 
        print("7. Exit") 

        choice = input("Enter your choice: ") 

        if choice == '1': 
            self.visualize_daily_expenses() 
        elif choice == '2': 
            self.visualize_weekly_expenses() 
        elif choice == '3': 
            self.visualize_monthly_expenses() 
        elif choice == '4': 
            self.visualize_yearly_expenses()
        elif choice == '5': 
            self.visualize_pie_chart() 
        elif choice == '6': 
            self.display_tabular_view() 
        elif choice == '7': 
            print("Exiting...") 
            return 
        else: 
            print("Invalid choice. Please try again.") 

    def visualize_daily_expenses(self): 
        dates = sorted(set(expense['date'] for expense in self.expenses)) 
        for date in dates: 
            expenses_on_date = [expense for expense in self.expenses if expense['date'] == date] 
            categories = [expense['category'] for expense in expenses_on_date] 
            amounts = [expense['amount'] for expense in expenses_on_date] 
            plt.figure(figsize=(8, 6)) 
            plt.bar(categories, amounts, color='skyblue') 
            plt.xlabel('Expense Categories') 
            plt.ylabel('Amount') 
            plt.title(f'Daily Expense Distribution on {date}') 
            plt.xticks(rotation=45) 
            plt.show() 

    def visualize_weekly_expenses(self): 
        dates = sorted(set(expense['date'] for expense in self.expenses)) 
        start_date = min(dates) 
        end_date = max(dates) 
        delta = timedelta(days=7) 
        while start_date <= end_date: 
            weekly_expenses = {} 
            for i in range(7): 
                current_date = start_date + timedelta(days=i) 
                expenses_on_date = [expense for expense in self.expenses if expense['date'] == current_date] 
                for expense in expenses_on_date: 
                    if expense['category'] in weekly_expenses: 
                        weekly_expenses[expense['category']] += expense['amount'] 
                    else: 
                        weekly_expenses[expense['category']] = expense['amount'] 
            if weekly_expenses: 
                categories = list(weekly_expenses.keys()) 
                amounts = list(weekly_expenses.values()) 
                plt.figure(figsize=(8, 6)) 
                plt.bar(categories, amounts, color='skyblue') 
                plt.xlabel('Expense Categories') 
                plt.ylabel('Amount') 
                plt.title(f'Weekly Expense Distribution from {start_date} to {start_date + timedelta(days=6)}') 
                plt.xticks(rotation=45) 
                plt.show() 
            start_date += delta 

    def visualize_monthly_expenses(self): 
        months = sorted(set(expense['date'].replace(day=1) for expense in self.expenses)) 
        for month in months: 
            monthly_expenses = {} 
            for expense in self.expenses: 
                if expense['date'].replace(day=1) == month: 
                    category = expense['category'] 
                    amount = expense['amount'] 
                    if category in monthly_expenses: 
                        monthly_expenses[category] += amount 
                    else: 
                        monthly_expenses[category] = amount 
            if monthly_expenses: 
                categories = list(monthly_expenses.keys()) 
                amounts = list(monthly_expenses.values()) 
                plt.figure(figsize=(8, 6)) 
                plt.bar(categories, amounts, color='skyblue') 
                plt.xlabel('Expense Categories') 
                plt.ylabel('Amount') 
                plt.title(f'Monthly Expense Distribution for {month.strftime("%B %Y")}') 
                plt.xticks(rotation=45) 
                plt.show() 

    def visualize_yearly_expenses(self):
        years = sorted(set(expense['date'].year for expense in self.expenses))
        for year in years:
            yearly_expenses = {}
            for expense in self.expenses:
                if expense['date'].year == year:
                    category = expense['category']
                    amount = expense['amount']
                    if category in yearly_expenses:
                        yearly_expenses[category] += amount
                    else:
                        yearly_expenses[category] = amount
            if yearly_expenses:
                categories = list(yearly_expenses.keys())
                amounts = list(yearly_expenses.values())
                plt.figure(figsize=(8, 6))
                plt.bar(categories, amounts, color='skyblue')
                plt.xlabel('Expense Categories')
                plt.ylabel('Amount')
                plt.title(f'Yearly Expense Distribution for {year}')
                plt.xticks(rotation=45)
                plt.show()

    def visualize_pie_chart(self): 
        categories = set(expense['category'] for expense in self.expenses) 
        category_amount = {category: sum(expense['amount'] for expense in self.expenses if expense['category'] == category) for category in categories} 
        plt.figure(figsize=(8, 8)) 
        plt.pie(category_amount.values(), labels=category_amount.keys(), autopct='%1.1f%%', startangle=140) 
        plt.title('Expense Distribution by Category') 
        plt.axis('equal') 
        plt.show() 

    def display_tabular_view(self): 
        df = pd.DataFrame(self.expenses) 
        print(df) 

def main(): 
    expense_list = ExpenseList() 

    while True: 
        print("\nExpense List Creator\n") 
        print("1. Add Expense") 
        print("2. Visualize Expenses") 
        print("3. Exit") 

        choice = input("Enter your choice: ") 

        if choice == '1': 
            date_str = input("Enter expense date (YYYY-MM-DD): ") 
            try: 
                date = datetime.strptime(date_str, '%Y-%m-%d').date() 
            except ValueError: 
                print("Invalid date format. Please use YYYY-MM-DD.") 
                continue 

            while True: 
                category = input("Enter expense category: ") 
                if any(c.isalpha() for c in category): 
                    break 
                else: 
                    print("Category must contain at least one non-numeric character.") 
            amount = float(input("Enter expense amount: ")) 
            expense_list.add_expense(date, category, amount) 
            print("Expense added successfully!") 
        elif choice == '2': 
            if expense_list.expenses: 
                expense_list.visualize_expenses() 
            else: 
                print("No expenses added yet.") 
        elif choice == '3': 
            print("Exiting...") 
            break 
        else: 
            print("Invalid choice. Please try again.") 

if __name__ == "__main__": 
    main()
```
